```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data in pandas
df1 = pd.read_csv(r"C:\Users\DELL E6400\Downloads\movies.csv")

df1.head()
```

```python
#Look for missing data
df1.isnull().sum()
```

```python
#drop rows with missing values

df = df1.dropna(axis = 0)

df.isnull().sum()
```

```python
#check data types

print(df.dtypes)
```

```python
#change data types for specific columns

df[['budget','runtime','gross','votes']] = df[['budget','runtime','gross','votes']].astype('int64')

df.head()
```

```python
df.sort_values(by =['gross'], inplace= False, ascending=False)
```

```python
#check for number of duplicate rows

df.duplicated().sum()
```

```python
#Use scatter plot to check for correlation betweeen budget and gross income
plt.scatter(x='gross', y='budget', data=df, c=df.gross.values)
plt.ylabel('Budget for Film')
plt.xlabel('Gross Earnings')
plt.title('Budget for Film vs Gross Earnings ')
plt.show()
```

```python
#Use regplot to check for correlation betweeen budget and gross income

sns.regplot(x="gross", y="budget", data=df,line_kws={'color':'purple'})
```

```python
#stripplot showing gross income vs rating

sns.stripplot(x="rating", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns

df.corr()
```

```python
# Visual Correlation Matrix between all numeric columns

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for 'numeric' Movie Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#numerize and correlate all fields in the data frame

df_numerized = df.copy()

for col_name in df_numerized.columns:

    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

print(df_numerized)

df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for  Movie Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#Movie columns with correlation >0.5

corr_pairs = df_numerized.corr().unstack()

#print(corr_pairs)

sorted_pairs = corr_pairs.sort_values(kind="quicksort")

#print(sorted_pairs)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Looking at the top 15 companies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:10].astype('int64')
 
CompanyGrossSumSorted
```

```python
#Barplot showing Gross revenue earned vs Company

CompanyGrossSumSorted.plot(kind='bar',color=(0.2, 0.4, 0.6, 0.6),figsize=(16, 4))

plt.title('Gross revenue earned vs Company')

plt.ylabel('Gross revenue earned')

plt.show()
```
